```python
# Ejercicio 1: Crea un array y reorganízalo en una matriz

def list_to_matrix_python(list_a, columns, rows):
    if columns < 0 or rows < 0:
        raise Exception("Columns and rows cannot be negative")
    elif columns*rows != len(list_a):
        raise Exception("Matrix dimensions do not match")

    return np.array([[list_a[j*columns + i] for i in range(columns)] for j in range(rows)], dtype=float)

    """
    Another way of doing the same more explicitly
    result = []
    for j in range(rows):
        row = []
        for i in range(columns):
           row.append(list_a[j*columns + i])
        result.append(row)

    return np.array(result)
    """

def list_to_matrix_numpy(list_a, columns, rows):
    arr = np.array(list_a, dtype=float)
    if arr.size != rows * columns:
        raise ValueError("Matrix dimensions do not match")
    return arr.reshape(rows, columns)

a = np.arange(1,21)  # No incluye el límite derecho
a_matrix = list_to_matrix_python(a, columns = 10, rows = 2)

print(a_matrix.shape)
print(a_matrix.ndim)
print(a_matrix.dtype)
print(a_matrix.size)

a_matrix
```

```python
# Ejercicio 2: Indexing y slicing.

# Dada la matriz:

A = np.arange(1, 17).reshape(4,4)

#Extrae:
# La segunda fila
# La tercera columna
# La submatriz central 2x2
# invierte las filas (de abajo hacia arriba)

# Recuerda que el slicing en numpy tine la siguiente estructura [start:stop:step]
# Para matrices: [fila, columna]

print("segunda fila: ", A[1,:])
print("Tercera columna ", A[:,2])
print("submatriz central 2x2: ")
print(A[1:3,1:3])

def inverse_rows(matrix_a):
    """Modifica la matriz original y la devuelve con sus filas invertidas
    (provoca efectos secundarios)."""
    for j in range(int(matrix_a.shape[0]/2)):
        temp = [n for n in matrix_a[j,:]]
        matrix_a[j,:] = matrix_a[-(j+1), :]
        matrix_a[-(j+1), :] = temp
    return matrix_a

print("\nMatriz con filas invertidas: ")
print(inverse_rows(A))

# Usando numpy las filas se invierten en una sola línea
def inverse_rows_numpy(matrix_a):
    """Regresa una copia de la matriz original con sus filas invertidas sin
    modificar la original (No provoca efectos secundarios)."""
    return matrix_a[::-1].copy()

print("\nInversión de filas con numpy: ")
print(inverse_rows_numpy(np.arange(1, 17).reshape(4,4)))
```

```python
# Ejercicio 3: Operaciones vectorizadas

x = np.linspace(0, 2*np.pi, 1000000)

# Calcula
# sin(x)
# cos(x)
# sin²(x) + cos²(x)

y = np.sin(x)**2 + np.cos(x)**2

# Error respecto a 1
error = np.abs(y - 1)

print("Error máximo:", np.max(error))
print("Error medio:", np.mean(error))
print(np.allclose(y, 1))

# Gráficas
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.plot(x, y, label="sin² + cos²")
plt.legend()
plt.show()
```
